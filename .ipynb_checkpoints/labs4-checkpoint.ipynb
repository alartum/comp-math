
# Transfer equation
## Formulation

$$
\begin{cases}
\frac{\partial u}{\partial t} + \frac{\partial u}{\partial x} = 0, \;\; 0 \le x \le 1, \;\; 0 \le t \le 0.5\\
u(x, 0) = g(x), \;\; 0 \le x \le 1 \\
u(0, t) = 1 \\
\end{cases}
$$

$$
g(x) =
\begin{cases}
0, & x > 0.2,\\
1, & x \le 0.2\\
\end{cases}
$$

## Numerical solution

$$\begin{cases}
u_0^{n+1} = u_1^{n+1} = 1, \\
u_m^{n+1} = (1-\sigma)u_m^n + \sigma u_{m-1}^n - \frac{\sigma(\sigma-1)}{2}(f(u_{m-1}^n, u_{m}^n, u_{m+1}^n) - f(u_{m-2}^n, u_{m-1}^n, u_{m}^n)) \\
\frac{u_M^{n+1} - u_M^{n}}{\tau} + \frac{u_M^{n} - u_{M-1}^{n}}{h} = 0
\end{cases}$$

$$
M = 100, \quad h = \frac{1}{M}, \quad \tau = \frac{h}{2}, \quad \sigma = \frac{\tau}{h} = \frac{1}{2}
$$

## Precise solution

$$
u(x, t) = g(x - ct)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.rcParams["animation.html"] = "html5"
np.set_printoptions(suppress=True, precision=3)
```

```python
def minmod(a, b):
    c = np.zeros_like(a)
    pos = np.logical_and(a > 0, b > 0)
    neg = np.logical_and(a < 0, b < 0)
    if np.any(pos):
        c[pos] = np.min((a[pos], b[pos]), axis=0)
    if np.any(neg):
        c[neg] = np.max((a[neg], b[neg]), axis=0)
    
    return c
    
def maxmod(a, b):
    c = np.zeros_like(a)
    pos = np.logical_and(a > 0, b > 0)
    neg = np.logical_and(a < 0, b < 0)
    if np.any(pos):
        c[pos] = np.max((a[pos], b[pos]), axis=0)
    if np.any(neg):
        c[neg] = np.min((a[neg], b[neg]), axis=0)
    
    return c
    
def left_triangle(x, y, z):
    return np.zeros_like(x)
    
def lax_wendroff(x, y, z):
    return z - y

def fromm(x, y, z):
    return (z - x)/2

def TVD_minmod(x, y, z):
    return minmod(z - y, y - x)

def TVD_superbee(x, y, z):
    return maxmod(minmod(2*(z-y), y-x), minmod(z-y, 2*(y-x)))

def TVD_Leer(x, y, z):
    return (np.abs(z-y)*(y-x) + np.abs(y-x)*(z-y))/\
           (np.abs(z-y) + np.abs(y-x) + 1e-14)
```

```python
T = 0.5
M = 100
h = 1/M
dt = h/2
s = dt/h

def tick(u, f, s):
    # Changed '+' to '-' here:          V
    #                                   V
#     u_new = (1-s)*u + s*np.roll(u, 1) + s*(1-s)/2*\
#             (f(np.roll(u, 1), u, np.roll(u, -1)) -\
#              f(np.roll(u, 2), np.roll(u, 1), u))
    u_new = (1-s)*u + s*np.roll(u, 1) - s*(1-s)/2*\
            (f(np.roll(u, 1), u, np.roll(u, -1)) -\
             f(np.roll(u, 2), np.roll(u, 1), u))
    u_new[0] = u_new[1] = 1
    u_new[-1] = (1-s)*u[-1] + s*u[-2]
    
    return u_new

def g(x):
    g = np.empty_like(x)
    g[x > 0.2] = 0
    g[x <= 0.2] = 1
    
    return g

def u_pr(x, t):
    return g(x - t)

xs = np.arange(0, 1+h/2, h)
tests = {'Left Triangle': left_triangle,
         'TVD minmod': TVD_minmod,
         'TVD superbee': TVD_superbee,
         'Lax-Wendroff': lax_wendroff,
         'Fromm': fromm,
         'TVD van-Leer': TVD_Leer}
us = {}
for pair in tests.items(): 
    us[pair[0]] = g(xs)

t = 0

while t < T + dt/2:
    for pair in tests.items(): 
        us[pair[0]] = tick(us[pair[0]], pair[1], s)
    t += dt

fig, ax = plt.subplots(1, 1, figsize=(10, 5))

for pair in tests.items():
    ax.plot(xs, us[pair[0]], label=pair[0])

u_0 = u_pr(xs, t)    
ax.plot(xs, u_0, label='Precise')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Transfer equation')
ax.axhline(c='k')
ax.axvline(x=0, c='k')
ax.axvline(x=1, c='k')
ax.grid()
ax.legend(loc='upper right')

plt.show()
```

# Time-independent Schrödinger equation

## Formulation

$$
\begin{cases}
\frac{\partial \Psi}{\partial t}  = \frac{i}{200} \frac{\partial^2 \Psi}{\partial x^2}, \;\; 0 \le x \le 1, \;\; 0 \le t \le 1.3\\
\Psi(x, 0) = \exp\left\{100ix - \frac{(x-0.3)^2}{0.01}\right\} \\
\Psi(0, t) = \Psi(1, t) = 0 \\
\end{cases}
$$

## Numerical solution

$$\begin{cases}
\frac{\Psi_m^{n+1} - \Psi_m^{n}}{\tau} = \frac{i}{400}\left[\frac{\Psi_{m-1}^{n+1} - 2\Psi_{m}^{n+1} + \Psi_{m+1}^{n+1}}{h^2} + \frac{\Psi_{m-1}^{n} -2\Psi_{m}^{n} + \Psi_{m+1}^{n}}{h^2}\right]\\
\Psi_{0}^{n+1} = \Psi_{M}^{n+1} = 0
\end{cases}$$

$$
M = 1000, \quad h = \frac{1}{M}, \quad \tau = h
$$


```python
%%capture --no-stderr --no-stdout
from scipy.linalg import solve_banded

T = 50
M = 1000
# Display each 'period'th iteration on animation
period = 50
h = 1/M
dt = h
s = dt/h**2

def tick(u, s):
    ab = np.ones((3, u.size), dtype=complex)
    ab[1, 1:-1] = 400j/s - 2
    ab[0, :2] = 0
    ab[2, -2:] = 0

    b = -np.roll(u, 1)+(2+400j/s)*u - np.roll(u, -1)
    b[0] = 0
    b[-1] = 0
    
    return solve_banded((1, 1), ab, b)

def psi(x):
    return np.exp(100j*x - (x-0.3)**2/0.01)

xs = np.arange(0, 1+h/2, h)
us = psi(xs)

t = 0
n = 0
ticks = []

while t < T + dt/2:
    us = tick(us, s)
    if not n % period:
        ticks.append(us.copy())
    n += 1
    t += dt

print("%d frames total" % len(ticks))
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

prob, = ax.plot(xs, np.abs(us)**2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Squared Wave Function Absolute Value')
ax.set_ylim(-2e-2, 2.5)
ax.axhline(c='k')
ax.axvline(x=0, c='k')
ax.axvline(x=1, c='k')
ax.grid()

def init():
    return prob,

def update(frame, xs, ticks):
    prob.set_data(xs, np.abs(ticks[frame])**2)
    return prob,

anim = FuncAnimation(fig, update, fargs=(xs, ticks), frames=len(ticks), 
                     interval=70, init_func=init, blit=True)

plt.show()
```
