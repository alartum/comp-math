## Определение положения по координатам

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y, z, t = sp.symbols('x y z t')

r_1 = [1, 1, 1, 2.00]
r_2 = [-1, -1, 1, 2.25]
r_3 = [-1, 1, -1, 1.70]
r_4 = [1, -1, -1, 1.50]

r_init = [0, 0, 0, 0]

F = sp.Matrix([(x-r_1[0])**2 + (y-r_1[1])**2 + (z-r_1[2])**2 - (t-r_1[3])**2,
               (x-r_2[0])**2 + (y-r_2[1])**2 + (z-r_2[2])**2 - (t-r_2[3])**2,
               (x-r_3[0])**2 + (y-r_3[1])**2 + (z-r_3[2])**2 - (t-r_3[3])**2,
               (x-r_4[0])**2 + (y-r_4[1])**2 + (z-r_4[2])**2 - (t-r_4[3])**2])

X = sp.Matrix([x, y, z, t])
J = F.jacobian(X)

J_sc = sp.lambdify((x, y, z, t), J, [{'ImmutableMatrix': np.array}, "numpy"])
def J(r):
    return J_sc(r[0], r[1], r[2], r[3])

F_sc = sp.lambdify((x, y, z, t), F, [{'ImmutableMatrix': np.array}, "numpy"])
def F(r):
    return F_sc(r[0], r[1], r[2], r[3])

r = np.array(r_init, dtype=float)
eps = 1e-14
max_it = 10

errs = []
for it in range(max_it):
    delta_r = np.linalg.solve(J(r), -F(r))
    r += delta_r.T[0]
    errs.append(np.linalg.norm(delta_r))
    if (np.linalg.norm(delta_r) < eps): 
        break

print('Решение:', r)
print('Ошибка (норма невязки) err =', errs[-1])
plt.title('Ошибка в зависимости от числа итераций')
plt.xlabel('Итерация')
plt.ylabel('Ошибка')
plt.plot(errs)
plt.show()
```

## Сплайны

Граничные условия на первую производную:

$$ f'(0) = f'(2 \pi) = 1$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve_banded

def f(x):
    return np.sin(x)

def build_spline(a, b, N, f, df_0, df_n):
    # Равномерная сетка
    xs, h = np.linspace(a, b, N, retstep=True)

    ab = np.zeros((3, N))
    ab[0, 2:]   = 1/h
    ab[1, 1:-1] = 4/h
    ab[1, 0] = ab[1, -1] = 1
    ab[2,  :-2] = 1/h
    # Производные на концах равны 0 и 1 соответственно
    b = np.zeros(N)
    b[0] = df_0
    b[-1] = df_n
    fs = f(xs)
    df = np.diff(fs)
    b[1:-1] = 3/h*(df[:-1] + df[1:])
    # Матрица не симметрична, через solveh_banded не выйдет
    return solve_banded((1, 1), ab, b), fs, xs

#Копипаста
def hermite(f1, m1, f2, m2, x1, x2, x):
    h = x2 - x1
    fd = (f2-f1)/h
    dx = x - x2
    
    return f2+m2*dx+(-3*fd+m1+2*m2)*dx**2/h + (m1+m2-2*fd)*dx**3/h**2

x_plt = np.arange(0, 2*np.pi, 10e-2)
plt.plot(x_plt, f(x_plt), 'r')

a, b = 0, 2*np.pi
df_0 = 1
df_n = 1
N = 3
m, fs, xs = build_spline(a, b, N, f, df_0, df_n)
err = 0
for i in range(1, len(m)):
    xs_local = np.linspace(xs[i-1], xs[i])
    spline = hermite(fs[i-1], m[i-1], fs[i], m[i], xs[i-1], xs[i], xs_local)
    err = max(np.max(np.fabs(spline - f(xs_local))), err)
    plt.plot(xs_local, spline, 'g')
plt.title(r'Приближение $sin x$ сплайном')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xs, fs, 'g.', ms=10)
print('Число точек N =', N)
print('Максимальное отклонение err =', err)

plt.show()
```

```python
a, b = 0, 2*np.pi
N = 5
df_0 = 1
df_n = 1

def test_spline(a, b, N, f, df_0, df_n):
    m, fs, xs = build_spline(a, b, N, f, df_0, df_n)
    err = 0
    for i in range(1, len(m)):
        xs_local = np.linspace(xs[i-1], xs[i])
        spline = hermite(fs[i-1], m[i-1], fs[i], m[i], xs[i-1], xs[i], xs_local)
        err = max(np.max(np.fabs(spline - f(xs_local))), err)

    return err

Ns = np.arange (100, 600, 10)
errs = [test_spline(a, b, N, f, df_0, df_n) for N in Ns]
log_errs = np.log(errs)

hs = 2*np.pi/(Ns-1)
plt.title('Зависимость ошибки от мелкости сетки')
plt.xlabel('Шаг')
plt.ylabel('Максимальное отклонение')
plt.plot(hs, errs, 'g.')
a, b = np.polyfit(hs, errs, deg = 1)
plt.plot(hs, a*hs + b)

print ('Линейная зависимость: err ~ {:.2}*h'.format(a))
plt.show()
```

## Интегрирование

$$\int_{0}^{\frac{\pi}{2}} \ln \frac{1}{\sin x} dx$$

В окрестности нуля:
$$ \ln \frac{1}{\sin x} = \ln \frac{1}{x} + \frac{x^2}{6} + O(x^2)$$

В окрестности $\frac{\pi}{2}$ регулярна.
Так что будем интегрировать следующую функцию:
$$  \ln \frac{1}{\sin x} - \ln \frac{1}{x} = \ln \frac{x}{\sin x}$$

Возьмём аналитически:

$$ \int_{0}^{\frac{\pi}{2}} \ln \frac{1}{x} dx = x  \ln \frac{1}{x} \bigg\vert_{0}^{\frac{\pi}{2}} + \int_{0}^{\frac{\pi}{2}} \frac{1}{x} x dx = \frac{\pi}{2}\ln\frac{2}{\pi} - x  \ln \frac{1}{x} \bigg\vert_{0} + \frac{\pi}{2}$$

Так как $\forall x > 0 \;\; \ln x < \sqrt{x}$, то
$$ \lim_{x \to 0} x \ln \frac{1}{x} \le \lim_{x \to 0} x \frac{1}{\sqrt{x}} = 0$$

Окончательно:
$$ \int_{0}^{\frac{\pi}{2}} \ln \frac{1}{x} dx = \frac{\pi}{2}\ln\frac{2}{\pi} + \frac{\pi}{2} = -\frac{\pi}{2} (-1 + \ln\frac{\pi}{2} )$$

Интеграл $\int_{0}^{\frac{\pi}{2}} \ln \frac{x}{\sin x} dx$ берём численно.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def gauss(f, x1, x2):
    a = (x2 + x1)/2
    b = (x2 - x1)/(2*np.sqrt(3))
    return (x2 - x1)/2 * (f(a + b) + f(a - b))

def f(x):
    return np.log(x/np.sin(x))

# Точное значение всего интеграла
I_precise = np.pi/2 * np.log(2)
# Часть, посчитанная аналитически
I_0 = -np.pi/2*(-1 + np.log(np.pi/2))
a, b = 0, np.pi/2
# Начальное разбиение
N = 10
eps = 1e-10

def runge(f, a, b, eps, N):
    I = 0
    dI = 0
    while True:
        xs = np.linspace(a, b, N)
        # Не включаем левый конец
        old_I = I
        I = 0
        for i in range(1, N-1):
            I += gauss(f, xs[i], xs[i+1])
        # У гаусса четвёртый порядок
        dI = np.fabs(I - old_I)/15
        if dI < eps: break
        N *= 2
    return I, dI

I, dI = runge(f, a, b, eps, N)
print('Оценка на ошибку: eps = ', dI)
print('Реальное значение ошибки: |I - I_precise| = ', np.abs(I + I_0 - I_precise))
```

## Дифференциальные уравнения

$$ \begin{cases}
   \frac{dy(t)}{dt} = -2ty(t) + 2t\\
   y(0) = 1
   \end{cases}$$
   
$$ y(t) = 1 + g(t)$$
$$  \frac{dg(t)}{dt} = -2tf(t) - 2t + 2t$$
$$ \frac{dg(t)}{dt} = -2tf(t) \Rightarrow g(t) = Ae^{-2t}$$


$$ y(t) = 1 + Ae^{-t^2}, \;\;\; y(0) = 1 + A = 1 \; \Rightarrow A = 0 \; \Rightarrow y(t) = 1$$

Если взять $y(0) = 2$, будет интереснее: $y(t) = 1 + e^{-t^2}$

Предлагается использовать метод Адамса:

$$\begin{cases}
\frac{u_{n+1} - u_n}{\tau} = \frac{3f(t_n, u_n) - f(t_{n-1}, u_{n-1})}{2}, \;\; n = 1, \dotsc, N - 1 \\
u_0 = y_0 \\
u_1 = y_0 + \tau f(0, y_0)
\end{cases}$$

```python
import matplotlib.pyplot as plt
import numpy as np

a, b = 0, 1
N = 20
y_0 = 2

def f(t, y):
    return 2*t*(1 - y)

def f_precise(t):
    return 1 + np.exp(-t**2)

ts, h = np.linspace(a, b, N, retstep = True)
us = np.zeros_like(ts)

us[0] = y_0
us[1] = y_0 + h * f(0, y_0)

for i in range(1, N-1):
    us[i+1] = us[i] + h/2*(3*f(ts[i], us[i]) - f(ts[i-1], us[i-1]))
    
plt.title('Сравнение точной и приближённой траекторий')
plt.plot(ts, us, 'r.')
plt.plot(ts, f_precise(ts), 'b')
plt.show()
```

```python
a, b = 0, 1
y_0 = 2

def calc_err(a, b, f, y_0, N, f_precise):
    ts, h = np.linspace(a, b, N, retstep = True)
    us = np.zeros_like(ts)

    us[0] = y_0
    us[1] = y_0 + h * f(0, y_0)

    for i in range(1, N-1):
        us[i+1] = us[i] + h/2*(3*f(ts[i], us[i]) - f(ts[i-1], us[i-1]))
    return np.max(np.fabs(us - f_precise(ts)))

Ns = np.arange(5, 30)
hs = (b-a)/(Ns-1)
errs = [calc_err(a, b, f, y_0, N, f_precise) for N in Ns]

plt.title('Величина ошибки в зависимости от шага')
plt.xlabel('Шаг')
plt.ylabel('Отклонение от точного решения')
plt.plot(hs, errs, 'r.')

a, b, c = np.polyfit(hs, errs, deg = 2)
print('err ~ ({:.2})h^2 + ({:.2})h + ({:.2})'.format(a, b, c))
approx = plt.plot(hs, a*hs**2 + b*hs + c, 'b')

plt.show()
```
